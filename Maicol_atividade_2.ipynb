## Gráfico da Função Gaussiana

A **Função Gaussiana** é uma importante função de densidade de probabilidade. Ela tem a forma de um sino com a boca virado para baixo segundo a expressão:
$$
H(x) = \left(\frac{1}{\sigma \sqrt{2\pi}}\right)e^{-\frac{1}{2}\left(\frac{x - \mu}{\sigma} \right)^2}
$$

onde $\mu$ é o seu valor médio e $\sigma$ é o desvio padrão da distribuição.

No primeiro gráfico, mostramos duas funções gaussianas com diferentes valores para $\mu$ e $\sigma$:
- A primeira com centro $\mu = 0$ e largura $\sigma = 1$ (em azul).
- A segunda com centro $\mu = 2$ e largura $\sigma = 2$ (em vermelho).


Esta fórmula descreve uma distribuição normal ou gaussiana, onde o parâmetro $a = \frac{1}{\sigma \sqrt{2\pi}}$ define a altura do pico, $\mu$ move o centro da curva ao longo do eixo $x$, e $\sigma$ afeta a largura da curva (quanto maior $\sigma$, mais espalhada a curva será).

Fonte : http://sampa.if.usp.br/~suaide/blog/?e=245

```python
def gaussian(x, sigma, mu):
    '''
    Calcula a função gaussiana.

    A função gaussiana é definida pela fórmula:
        f(x) = (1 / (sigma * sqrt(2 * pi))) * exp(-0.5 * ((x - mu) / sigma) ** 2)
    onde sigma é o desvio padrão e mu é o valor médio da distribuição.

    Argumentos
    ----------
    x (array): Valores de entrada para calcular a função gaussiana.

    sigma (float): Desvio padrão da distribuição (determina a largura da curva).

    mu (float): Valor médio da distribuição (determina o centro da curva).

    Retorna
    -------
    array: Valores calculados da função gaussiana para cada valor de x.
    '''
    # Amplitude da função gaussiana
    a = 1 / (sigma * np.sqrt(2 * np.pi))  # Amplitude da função gaussiana

    # Calcula a função gaussiana usando a fórmula fornecida
    return a * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Função para plotar todas as funções gaussianas no mesmo gráfico
def all_gauss(params):
    '''
    Plota múltiplas funções gaussianas no mesmo gráfico.

    Argumentos
    ----------
    params (tupla): Lista de tuplas com os parâmetros (sigma, mu, cor) de cada função gaussiana.

    Retorna
    -------
    None : Plota o gráfico das funções gaussianas.
    '''
    x = np.linspace(-12, 12, 1000)
    plt.figure(figsize=(10, 6))

    for sigma, mu, color in params:
        y = gaussian(x, sigma, mu)
        plt.plot(x, y, label=f'μ={mu}, σ={sigma}', color=color)

    plt.title('Funções Gaussianas') # Adiciona título ao gráfico
    plt.legend()  # Adiciona legenda ao gráfico
    plt.grid(True)  # Adiciona grade ao gráfico
    plt.xlabel('x')  # Adiciona rótulo ao eixo x
    plt.ylabel('H(x)')  # Adiciona rótulo ao eixo y
    plt.show()  # Exibe o gráfico

# Definindo os parâmetros e cores para diferentes funções gaussianas
params = [
    (4, 0, 'red'),    # sigma = 4, mu = 0
    (3, 1, 'green'),    # sigma = 3, mu = 1
    (2, 2, 'blue')       # sigma = 2, mu = 2
]

# Plotando todas as funções gaussianas no mesmo gráfico
all_gauss(params)
```

## Gráfico da Função Sinc

A **Função Sinc** é dada por:
$$
\text{sinc}(x) = \frac{\sin(x)}{x}
$$
para $x \neq 0$, e sinc$(0) = 1$.

No gráfico da função sinc$(x)$, podemos ver que a função oscila e decai conforme o valor de $x$ aumenta em magnitude. Esse tipo de função é muito importante em processamento de sinais e teoria da informação, devido às suas propriedades de interpolação e reconstrução de sinais.


```python
# Função Sinc definida para todos os valores de x
def sinc(x):
    '''
    Calcula a função sinc(x).

    A função sinc é definida como sin(x)/x para x diferente de 0 e 1 para x igual a 0.
    É amplamente utilizada em processamento de sinais e teoria da informação devido às suas propriedades de interpolação e reconstrução de sinais.

    Argumentos
    ----------
    x (float ou array): Os valores para os quais a função sinc será calculada.

    Retorna
    -------
    float ou array: Os valores da função sinc calculados para os valores de x.
    '''
    return np.where(x != 0, np.sin(x) / x, 1)

# Intervalo de valores de x
x = np.linspace(-20, 20, 1000)

# Calcula os valores da função sinc
y = sinc(x)

# Criando o plot
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Função Sinc') # Adiciona título ao gráfico
plt.grid(True)  # Adiciona grade ao gráfico
plt.xlabel('x')  # Adiciona rótulo ao eixo x
plt.ylabel('sinc(x)')  # Adiciona rótulo ao eixo y
plt.show()  # Exibe o gráfico
```

## Gráfico das Funções Exponenciais

A **Função Exponencial** é dada por:
$$
f(x) = e^{ax}
$$
onde $a$ é uma constante.

No gráfico, mostramos quatro funções exponenciais com diferentes valores de $a$:
- $a = -10$ (em roxo), que decai rapidamente para zero.
- $a = -1$ (em laranja), que também decai, mas de forma mais suave.
- $a = 2$ (em rosa), que cresce exponencialmente.
- $a = 10$ (em marrom), que cresce muito rapidamente.

Essas funções são comumente usadas para modelar crescimento ou decaimento exponencial, como o crescimento populacional ou a radioatividade.


```python
def exp(x, a):
    '''
    Calcula a função exponencial.

    A função exponencial é definida pela fórmula: f(x) = exp(a * x)
    onde a é uma constante que determina a taxa de crescimento ou decaimento.

    Argumentos
    ----------
    x (array): Valores de entrada para calcular a função exponencial.
    a (float): Constante que determina a taxa de crescimento ou decaimento da função.

    Retorna
    -------
    array :Valores calculados da função exponencial para cada valor de x.
    '''
    return np.exp(a * x)

def all_exp(params):
    '''
    Plota várias funções exponenciais com diferentes parâmetros.

    Argumentos
    ----------
    params (tupla): Lista de tuplas contendo os valores de 'a' e as cores correspondentes para cada função exponencial.

    Retorna
    -------
    None : Plota o gráfico das funções exponenciais.
    '''
    plt.figure(figsize=(10, 6))
    x = np.linspace(-2, 2, 1000)  # Gera 1000 valores igualmente espaçados entre -2 e 2
    y_max = 4  # Definir limite do eixo y

    # Plotando cada função exponencial
    for a, color in params:
        y = exp(x, a)
        y = np.clip(y, None, y_max)  # Limita o valor máximo de y a 4
        plt.plot(x, y, label=f'a={a}', color=color)

    # Configurações do gráfico
    plt.title('Funções Exponenciais')  # Adiciona título ao gráfico
    plt.legend()  # Adiciona legenda ao gráfico
    plt.grid(True)  # Adiciona grade ao gráfico
    plt.xlabel('x')  # Adiciona rótulo ao eixo x
    plt.ylabel('f(x) = exp(ax)')  # Adiciona rótulo ao eixo y
    plt.ylim([0, y_max])  # Limita o eixo y
    plt.show()  # Exibe o gráfico

# Definindo os parâmetros para diferentes funções exponenciais
params = [
    (-10, 'purple'),    # a = -10
    (-1, 'orange'),     # a = -1
    (2, 'pink'),        # a = 2
    (10, 'brown'),      # a = 10
]

# Chamando a função all_exp para plotar as funções exponenciais
all_exp(params)
```

## Gráfico da Funções Quadráticas

A **Função Quadrática** é dada por:
$$
f(x) = ax^2 + bx + c
$$
onde $a$, $b$ e $c$ são constantes.

No gráfico, apresentamos três situações diferentes de funções quadráticas e uma situação com função linear.


A primeira função quadrática com os coeficientes $a = 1$, $b = -5$ e $c = 6$. As raízes da função são encontradas utilizando a fórmula de Bhaskara. Para esta função, temos duas raízes reais, que são mostradas no gráfico como os pontos onde a curva cruza o eixo $x$.

que possui um discriminante igual a zero.
$$\Delta = b^2 - 4ac > 0$$

As raízes são aproximadamente $x_1 = 2$ e $x_2 = 3$, e o gráfico exibe as interseções da parábola com o eixo $x$ (em vermelho).

A segunda função quadrática quadráticas, consideramos a equação:
$$
f(x) = x^2 -2x + 1
$$
que possui um discriminante igual a zero.
$$\Delta = b^2 - 4ac = 0$$

Quando o discriminante é zero, a função tem uma única raiz real. Neste caso, a raiz é $x = 1$, mostrada no gráfico. A parábola toca o eixo $x$ em um único ponto, representando a raiz única da equação.

A terceira função quadrática, temos a equação:
$$
f(x) = x^2 + 2x + 3
$$
que possui um discriminante negativo.$$\Delta = b^2 - 4ac < 0$$

Quando o discriminante é negativo, a equação não possui raízes reais. Em vez disso, as raízes são complexas. O gráfico mostra uma parábola que não toca o eixo $x$, indicando que não existem soluções reais para essa equação. As raízes complexas são representadas como números imaginários.

A quarta função é linear, onde usamos a equação:
$$
f(x) = 0\cdot x^2 + 2x + 6
$$

Por ser uma equação linear, no gráfico vemos uma reta em vez de uma parábola como nas funções quadráticas.

```python
def bhaskara(a, b, c):
    '''
    Calcula as raízes de uma equação quadrática do tipo ax² + bx + c = 0.

    A função calcula as raízes reais de uma equação quadrática. Se o coeficiente 'a' for zero,
    a equação se torna linear e a função retorna duas raízes iguais. Se o discriminante (delta) for negativo,
    a função retorna duas raízes com parte real e imaginária, resultando em raízes complexas

    Argumentos:
    a (float): Coeficiente da variável x².
    b (float): Coeficiente da variável x.
    c (float): Termo constante.

    Retorna:
    tuple: Uma tupla com as duas raízes, que podem ser reais ou complexas:
           - Se delta >= 0, retorna duas raízes reais.
           - Se delta < 0, retorna as raízes complexas.
           - Se a == 0, retorna duas raízes iguais.

    Verifica condições
    -------------------
    Se (a) for zero e a equação for linear, mas (b) também for zero (caso indeterminado), a função irá levantar uma exceção `ValueError`.

    '''

    # Verifica se 'a' é zero
    if a == 0:
        # Se a == 0, a equação não é quadrática, mas é uma equação linear.
        # Nesse caso, retorna uma raiz única, x = -c/b.
        if b == 0:
            raise ValueError("Não é possível resolver para a equação linear 0x + c = 0 quando c não é zero.")
        x = -c / b
        return (x, x)  # Duas raízes iguais

    # Cálculo do delta
    delta = b**2 - 4*a*c

    # Se delta for negativo, as raízes são complexas
    if delta < 0:
        real_part = -b / (2*a)
        imaginary_part = (-delta)**0.5 / (2*a)
        # Retorna as raízes complexas (parte real, parte imaginária)
        return (real_part + imaginary_part * 1j, real_part - imaginary_part * 1j)

    # Se delta for não-negativo, calcula as raízes reais
    x1 = (-b + (delta**0.5)) / (2*a)
    x2 = (-b - (delta**0.5)) / (2*a)
    x  = (x1, x2)

    # Retorna as duas raízes reais
    return x

# Função para plotar todas as funções quadráticas no mesmo gráfico
def all_quad(coefficients):
    '''
    Plota múltiplas funções quadráticas ax^2 + bx + c e indica as raízes com círculos.

    Argumentos
    ----------
    params (tupla): Lista de tuplas contendo os valores de 'a' e as cores correspondentes para cada função exponencial.

    Retorna
    -------
    None : Plota o gráfico das funções exponenciais.
    '''
    x = np.linspace(-17, 17, 400)
    plt.figure(figsize=(10, 6))

    for a, b, c, color in coefficients:
        y = a*x**2 + b*x + c
        roots = bhaskara(a, b, c)

        plt.plot(x, y, label=f'${a}x^2 + {b}x + {c}$', color=color)
        if roots[0] is not None and roots[1] is not None:
            plt.plot(roots[0], 0, 'ro')  # Marcando a primeira raiz com um círculo vermelho
            plt.plot(roots[1], 0, 'ro')  # Marcando a segunda raiz com um círculo vermelho


    plt.axhline(0, color='black', linewidth=0.5) # Adiciona uma linha horizontal no eixo y em y=0
    plt.axvline(0, color='black', linewidth=0.5) # Adiciona uma linha vertical no eixo x em x=0
    plt.title('Funções Quadráticas') # Adiciona título ao gráfico
    plt.legend()  # Adiciona legenda ao gráfico
    plt.grid(True)  # Adiciona grade ao gráfico
    plt.xlabel('x')  # Adiciona rótulo ao eixo x
    plt.ylabel('y') # Adiciona rótulo ao eixo y
    plt.ylim(-25, 200) # Define o intervalo do eixo y
    plt.show()  # Exibe o gráfico


# Definindo os coeficientes e cores para três casos
params = [
    (1, -5, 6, 'red')  ,    # Discriminante positivo
    (1, -2, 1, 'green'),    # Discriminante zero
    (1, 2, 3, 'blue')  ,    # Discriminante negativo
    (0, 2, 6, 'grey')       # coeficente a = 0
]

# Plotando todas as funções quadráticas no mesmo gráfico
all_quad(params)

# Exemplos de teste

# Caso com raízes reais
x1, x2 = bhaskara(1, -5, 6)
print('\nRaízes reais:')
print('Primeira raiz:', x1)
print('Segunda raiz:', x2)

# Caso descriminante = 0
x1, x2 = bhaskara(1, -2, 1)
print('\nDiscriminante zero:')
print('Raíz:', x1)

# Caso com raízes complexas
x1, x2 = bhaskara(1, 2, 3)
print('\nRaízes complexas:')
print('Primeira raiz:', x1)
print('Segunda raiz:', x2)

# Caso com coeficiente a = 0 (equação linear)
x1, x2 = bhaskara(0, 2, 6)
print('\nEquação linear:')
print('Raíz:', x1)
```

## Gráfico da Função Gaussiana por meio de arquivos .txt

```python
def gaussian(x, sigma, mu):
    '''
    Calcula a função gaussiana.

    A função gaussiana é definida pela fórmula:
        f(x) = (1 / (sigma * sqrt(2 * pi))) * exp(-0.5 * ((x - mu) / sigma) ** 2)
    onde sigma é o desvio padrão e mu é o valor médio da distribuição.

    Argumentos
    ----------
    x (array): Valores de entrada para calcular a função gaussiana.

    sigma (float): Desvio padrão da distribuição (determina a largura da curva).

    mu (float): Valor médio da distribuição (determina o centro da curva).

    Retorna
    -------
    array: Valores calculados da função gaussiana para cada valor de x.
    '''
    # Amplitude da função gaussiana
    a = 1 / (sigma * np.sqrt(2 * np.pi))  # Amplitude da função gaussiana

    # Calcula a função gaussiana usando a fórmula fornecida
    return a * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Definindo os parâmetros para diferentes funções gaussianas
params = [
    (4, 0),    # sigma = 4, mu = 0
    (3, 1),    # sigma = 3, mu = 1
    (2, 2)       # sigma = 2, mu = 2
]

# Função para plotar todas as funções gaussianas no mesmo gráfico
def all_gauss(params):
    '''
    Plota múltiplas funções gaussianas no mesmo gráfico.

    Argumentos
    ----------
    params (tupla): Lista de tuplas com os parâmetros (sigma, mu, cor) de cada função gaussiana.

    Retorna
    -------
    None : Plota o gráfico das funções gaussianas.
    '''
    x = np.linspace(-12, 12, 1000)

    for sigma, mu in params:
        y = gaussian(x, sigma, mu)

# Função para gerar os dados das funções gaussianas e salvar em arquivos
def generate_gaussian_data():
    '''
    Gera dados para funções gaussianas com diferentes valores de média (mu) e desvio padrão (sigma) e salva os resultados em arquivos de texto.

    A função cria um conjunto de valores de entrada `x` variando de -12 a 12, e para cada combinação de parâmetros `sigma` e `mu` fornecida em `params`,
    calcula os valores da função gaussiana correspondente. Os dados gerados são então salvos em arquivos de texto, onde cada arquivo contém duas colunas:
    - A primeira coluna contém os valores de `x`.
    - A segunda coluna contém os valores da função gaussiana `f(x)`.

    O nome do arquivo de saída é no formato 'gauss_mu{mu}_sigma{sigma}.txt', onde `{mu}` e `{sigma}` são os valores dos parâmetros.
    '''

    x = np.linspace(-12, 12, 1000)
    for sigma, mu in params:
        # This line was not properly indented before
        y = gaussian(x, sigma, mu)

        # Nome do arquivo de saída
        filename = f'gauss_mu{mu}_sigma{sigma}.txt'

        # Salvar os dados no arquivo
        np.savetxt(filename, np.column_stack((x, y)), header="x, f(x)", delimiter=",")
        print(f"Dados salvos em {filename}")

# Gerar e salvar os dados
generate_gaussian_data()
```

```python
# Função para ler e plotar os dados salvos
def plot_gaussian_data():
    """
    Lê os dados salvos de funções gaussianas a partir de arquivos de texto e plota os gráficos correspondentes.

    A função lê os arquivos de texto especificados na lista `files`, que contêm os valores de `x` e os valores calculados da função gaussiana `f(x)`
    para diferentes combinações de média (`mu`) e desvio padrão (`sigma`). Para cada arquivo, os dados são extraídos e a função gera um gráfico
    com as curvas gaussianas correspondentes. O gráfico exibe múltiplas curvas, uma para cada arquivo, e inclui legendas, título e rótulos nos eixos.

    Os arquivos de entrada seguem o formato 'gauss_mu{mu}_sigma{sigma}.txt', onde `{mu}` e `{sigma}` representam os parâmetros da função gaussiana.
    """

    # Lista de arquivos para ler
    files = [
        'gauss_mu0_sigma4.txt',   # Arquivo com sigma = 4, mu = 0
        'gauss_mu1_sigma3.txt',   # Arquivo com sigma = 3, mu = 1
        'gauss_mu2_sigma2.txt'    # Arquivo com sigma = 2, mu = 2
    ]

    # Criar o gráfico
    plt.figure(figsize=(10, 6))

    # Ler e plotar os dados de cada arquivo
    for file in files:
        data = np.loadtxt(file, delimiter=',', skiprows=1)
        x = data[:, 0]  # Primeira coluna: valores de x
        y = data[:, 1]  # Segunda coluna: valores de f(x)

        # Extraímos sigma e mu do nome do arquivo
        sigma = file.split('_')[2][5:]  # Extrai o valor de sigma do nome do arquivo
        mu = file.split('_')[1][2:]     # Extrai o valor de mu do nome do arquivo

        # Plotando os dados
        plt.plot(x, y, label=f'μ={mu}, σ={sigma}')

    plt.title('Funções Gaussianas') # Adiciona título ao gráfico
    plt.legend()  # Adiciona legenda ao gráfico
    plt.grid(True)  # Adiciona grade ao gráfico
    plt.xlabel('x')  # Adiciona rótulo ao eixo x
    plt.ylabel('H(x)')  # Adiciona rótulo ao eixo y
    plt.show()  # Exibe o gráfico

# Chama a função para plotar os dados
plot_gaussian_data()
```

## Gráfico da Função Sinc por meio de arquivo .txt

```python
# Função Sinc definida para todos os valores de x
def sinc(x):
    '''
    Calcula a função sinc(x).

    A função sinc é definida como sin(x)/x para x diferente de 0 e 1 para x igual a 0.
    É amplamente utilizada em processamento de sinais e teoria da informação devido às suas propriedades de interpolação e reconstrução de sinais.

    Argumentos
    ----------
    x (float ou array): Os valores para os quais a função sinc será calculada.

    Retorna
    -------
    float ou array: Os valores da função sinc calculados para os valores de x.
    '''
    return np.where(x != 0, np.sin(x) / x, 1)

# Função para gerar os dados da função sinc e salvar em um arquivo
def generate_sinc_data():
    '''
    Gera os dados da função sinc e salva os resultados em um arquivo de texto.

    A função calcula os valores da função sinc para um intervalo de valores de `x` variando de -20 a 20. Os dados gerados são então salvos em um arquivo
    chamado 'sinc_data.txt'. O arquivo contém duas colunas: a primeira coluna com os valores de `x` e a segunda com os valores de `sinc(x)`.

    O arquivo gerado segue o formato:
    - Primeira coluna: valores de `x`.
    - Segunda coluna: valores da função sinc `sinc(x)`.
    '''
    # Intervalo de valores de x
    x = np.linspace(-20, 20, 1000)

    # Calcula os valores da função sinc
    y = sinc(x)

    filename = 'sinc_data.txt'

    # Salvar os dados no arquivo
    np.savetxt(filename, np.column_stack((x, y)), header="x, sinc(x)", delimiter=",")
    print(f"Dados salvos em {filename}")

# Gerar e salvar os dados
generate_sinc_data()
```

```python
# Função para ler e plotar os dados da função sinc
def plot_sinc_data():
    '''
    Lê os dados da função sinc de um arquivo de texto e plota o gráfico correspondente.

    A função carrega os dados da função sinc a partir do arquivo 'sinc_data.txt', onde os valores de `x` e `sinc(x)` estão armazenados em duas colunas.
    O gráfico é gerado utilizando esses dados, mostrando a curva da função sinc ao longo do intervalo de valores de `x`. O gráfico inclui título,
    rótulos nos eixos e uma grade para melhor visualização.

    O arquivo de entrada 'sinc_data.txt' deve conter duas colunas:
    - A primeira coluna: valores de `x`.
    - A segunda coluna: valores da função sinc `sinc(x)`.
    '''
    # Carregar os dados do arquivo
    data = np.loadtxt('sinc_data.txt', delimiter=',', skiprows=1)

    # Separando os valores de x e sinc(x)
    x = data[:, 0]  # Primeira coluna: valores de x
    y = data[:, 1]  # Segunda coluna: valores de sinc(x)

    # Criando o gráfico
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title('Função Sinc') # Adiciona título ao gráfico
    plt.grid(True)  # Adiciona grade ao gráfico
    plt.xlabel('x')  # Adiciona rótulo ao eixo x
    plt.ylabel('sinc(x)')  # Adiciona rótulo ao eixo y
    plt.show()  # Exibe o gráfico

# Chama a função para plotar os dados
plot_sinc_data()
```

## Gráfico das Funções exponenciais por meio de arquivo .txt

```python
def exp(x, a):
    '''
    Calcula a função exponencial.

    A função exponencial é definida pela fórmula: f(x) = exp(a * x)
    onde a é uma constante que determina a taxa de crescimento ou decaimento.

    Argumentos
    ----------
    x (array): Valores de entrada para calcular a função exponencial.
    a (float): Constante que determina a taxa de crescimento ou decaimento da função.

    Retorna
    -------
    array :Valores calculados da função exponencial para cada valor de x.
    '''
    return np.exp(a * x)

# Função para gerar os dados das funções exponenciais e salvar em arquivos
def generate_exp_data():
    # Intervalo de valores de x
    x = np.linspace(-2, 2, 1000)  # Gera 1000 valores igualmente espaçados entre -2 e 2
    y_max = 4  # Definir limite do eixo y

    # Definindo os parâmetros para diferentes funções exponenciais
    params = [
        (-10, 'purple'),    # a = -10
        (-1, 'orange'),     # a = -1
        (2, 'pink'),        # a = 2
        (10, 'brown'),      # a = 10
    ]

    # Gerar e salvar os dados
    for a, color in params:
        y = exp(x, a)
        y = np.clip(y, None, y_max)  # Limita o valor máximo de y a 4

        # Nome do arquivo de saída
        filename = f'exp_data_{a}.txt'

        # Salvar os dados no arquivo
        np.savetxt(filename, np.column_stack((x, y)), header="x, f(x)", delimiter=",")
        print(f"Dados salvos em {filename}")

# Gerar e salvar os dados
generate_exp_data()
```

```python
# Função para ler e plotar os dados das funções exponenciais
def plot_exp_data():
    '''
    Lê os dados de arquivos de funções exponenciais e plota os gráficos correspondentes.

    A função percorre uma lista de arquivos de dados contendo valores da função exponencial
    no formato f(x) = exp(ax), onde 'a' é extraído do nome do arquivo. Para cada arquivo,
    os dados são lidos e plotados em um gráfico, onde o eixo x contém os valores de x e o eixo y
    contém os valores correspondentes de f(x). A legenda do gráfico inclui o valor de 'a' extraído
    do nome do arquivo
    '''
    # Lista de arquivos para ler
    files = [
        'exp_data_-10.txt',   # Arquivo com a = -10
        'exp_data_-1.txt',    # Arquivo com a = -1
        'exp_data_2.txt',     # Arquivo com a = 2
        'exp_data_10.txt'     # Arquivo com a = 10
    ]

    # Criar o gráfico
    plt.figure(figsize=(10, 6))
    y_max = 4

    # Ler e plotar os dados de cada arquivo
    for file in files:
        data = np.loadtxt(file, delimiter=',', skiprows=1)
        x = data[:, 0]  # Primeira coluna: valores de x
        y = data[:, 1]  # Segunda coluna: valores de f(x)

        # Extraímos o valor de 'a' do nome do arquivo
        a = file.split('_')[2][:-4] # Extrai o valor de 'a' do nome do arquivo

        # Plotando os dados
        plt.plot(x, y, label=f'a={a}')

    # Configurações do gráfico
    plt.title('Funções Exponenciais')  # Adiciona título ao gráfico
    plt.legend()  # Adiciona legenda ao gráfico
    plt.grid(True)  # Adiciona grade ao gráfico
    plt.xlabel('x')  # Adiciona rótulo ao eixo x
    plt.ylabel('f(x) = exp(ax)')  # Adiciona rótulo ao eixo y
    plt.ylim([0, y_max])  # Limita o eixo y
    plt.show()  # Exibe o gráfico

# Chama a função para plotar os dados
plot_exp_data()
```

## Gráfico da Funções Quadráticas por meio de arquivo .txt

```python
def bhaskara(a, b, c):
    '''
    Calcula as raízes de uma equação quadrática do tipo ax² + bx + c = 0.

    A função calcula as raízes reais de uma equação quadrática. Se o coeficiente 'a' for zero,
    a equação se torna linear e a função retorna duas raízes iguais. Se o discriminante (delta) for negativo,
    a função retorna duas raízes com parte real e imaginária, resultando em raízes complexas

    Argumentos:
    a (float): Coeficiente da variável x².
    b (float): Coeficiente da variável x.
    c (float): Termo constante.

    Retorna:
    tuple: Uma tupla com as duas raízes, que podem ser reais ou complexas:
           - Se delta >= 0, retorna duas raízes reais.
           - Se delta < 0, retorna as raízes complexas.
           - Se a == 0, retorna duas raízes iguais.

    Verifica condições
    -------------------
    Se (a) for zero e a equação for linear, mas (b) também for zero (caso indeterminado), a função irá levantar uma exceção `ValueError`.

    '''

    # Verifica se 'a' é zero
    if a == 0:
        # Se a == 0, a equação não é quadrática, mas é uma equação linear.
        # Nesse caso, retorna uma raiz única, x = -c/b.
        if b == 0:
            raise ValueError("Não é possível resolver para a equação linear 0x + c = 0 quando c não é zero.")
        x = -c / b
        return (x, x)  # Duas raízes iguais

    # Cálculo do delta
    delta = b**2 - 4*a*c

    # Se delta for negativo, as raízes são complexas
    if delta < 0:
        real_part = -b / (2*a)
        imaginary_part = (-delta)**0.5 / (2*a)
        # Retorna as raízes complexas (parte real, parte imaginária)
        return (real_part + imaginary_part * 1j, real_part - imaginary_part * 1j)

    # Se delta for não-negativo, calcula as raízes reais
    x1 = (-b + (delta**0.5)) / (2*a)
    x2 = (-b - (delta**0.5)) / (2*a)
    x  = (x1, x2)

    # Retorna as duas raízes reais
    return x


# Função para gerar os dados das funções quadráticas e salvar em arquivos
def generate_quad_data():
    # Intervalo de valores de x
    x = np.linspace(-17, 17, 400)
    coefficients = [
        (1, -5, 6, 'red'),  # Discriminante positivo
        (1, -2, 1, 'green'),  # Discriminante zero
        (1, 2, 3, 'blue'),  # Discriminante negativo
        (0, 2, 6, 'grey')  # coeficiente a = 0
    ]
    # Gerar e salvar os dados
    for a, b, c, color in coefficients:
        y = a*x**2 + b*x + c
        roots = bhaskara(a, b, c)

        # Preparando as raízes para salvar no arquivo
        if isinstance(roots[0], complex):  # Se as raízes forem complexas
            root1 = np.full_like(x, roots[0].real)  # Repetir a parte real
            root2 = np.full_like(x, roots[0].imag)  # Repetir a parte imaginária
        else:  # Se as raízes forem reais
            root1 = np.full_like(x, roots[0])  # Repetir a raiz 1
            root2 = np.full_like(x, roots[1])  # Repetir a raiz 2

        # Nome do arquivo de saída
        filename = f'quad_data_{a}_{b}_{c}.txt'

        # Salvar os dados no arquivo
        np.savetxt(filename, np.column_stack((x, y, root1, root2)),
                   header="x, f(x), root1, root2", delimiter=",")
        print(f"Dados salvos em {filename}")

# Gerar e salvar os dados
generate_quad_data()
```

```python
# Função para ler e plotar os dados das funções quadráticas
def plot_quadratic_data():
    # Lista de arquivos para ler
    files = [
        'quad_data_1_-5_6.txt',
        'quad_data_1_-2_1.txt',
        'quad_data_1_2_3.txt',
        'quad_data_0_2_6.txt'
    ]

    # Criar o gráfico
    plt.figure(figsize=(10, 6))

    # Ler e plotar os dados de cada arquivo
    for file in files:
        data = np.loadtxt(file, delimiter=',', skiprows=1)
        x = data[:, 0]  # Primeira coluna: valores de x
        y = data[:, 1]  # Segunda coluna: valores de f(x)
        root1 = data[0, 2]     # Terceira coluna: primeira raiz
        root2 = data[0, 3]     # Quarta coluna: segunda raiz

        # Plotando a função quadrática
        plt.plot(x, y, label=f'{file.split("_")[2]}x^2 + {file.split("_")[3]}x + {file.split("_")[4].split(".")[0]}')

        # Marcando as raízes
        plt.plot(root1, 0, 'ro')  # Marcando a primeira raiz com um círculo vermelho
        plt.plot(root2, 0, 'ro')  # Marcando a segunda raiz com um círculo vermelho

    plt.axhline(0, color='black', linewidth=0.5) # Adiciona uma linha horizontal no eixo y em y=0
    plt.axvline(0, color='black', linewidth=0.5) # Adiciona uma linha vertical no eixo x em x=0
    plt.title('Funções Quadráticas') # Adiciona título ao gráfico
    plt.legend()  # Adiciona legenda ao gráfico
    plt.grid(True)  # Adiciona grade ao gráfico
    plt.xlabel('x')  # Adiciona rótulo ao eixo x
    plt.ylabel('y') # Adiciona rótulo ao eixo y
    plt.ylim(-25, 200) # Define o intervalo do eixo y
    plt.show()  # Exibe o gráfico

# Chama a função para plotar os dados
plot_quadratic_data()
```

```python
"""
Este script plota o gráfico das curvas paramétricas de um deltoide.

O deltoide é uma curva definida por uma parametrização em termos do ângulo theta. A parametrização é dada por:

    x(θ) = 2 * cos(θ) + cos(2 * θ)
    y(θ) = 2 * sin(θ) - sin(2 * θ)

O código utiliza a biblioteca `numpy` para calcular os valores de x e y ao longo de um intervalo de 0 a 2π e a biblioteca `matplotlib` para exibir o gráfico.

Funções utilizadas:
- np.linspace: Gera uma sequência de valores para o parâmetro theta.
- np.cos e np.sin: Calculam os valores de x e y com base na parametrização do deltoide.
- plt.plot: Plota o gráfico das coordenadas (x, y).
- plt.xlabel, plt.ylabel, plt.title, plt.legend, plt.grid e plt.axis: Personalizam o gráfico, incluindo título, legendas e formato da grade.

O gráfico gerado é uma representação visual da curva do deltoide no plano cartesiano.

Bibliotecas requeridas:
- numpy (np)
- matplotlib.pyplot (plt)
"""

# Definindo o intervalo de teta
theta = np.linspace(0, 2 * np.pi, 1000)

# Calculando as coordenadas x e y
x = 2 * np.cos(theta) + np.cos(2 * theta)
y = 2 * np.sin(theta) - np.sin(2 * theta)

# Criando o gráfico
plt.figure()
plt.plot(x, y, label='Deltoide')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico das Curvas Paramétricas do Deltoide')
plt.legend()
plt.axis('equal')
plt.show()


#https://commons.m.wikimedia.org/wiki/File:Curva_deltoide.jpg
```
